# Tratando Outliers
Calculando os limites, pontos distantes da média

```python
#https://minerandodados.com.br/tratando-outliers-python/
#Calculando o desvio padrão dos gastos
mean = np.mean(df1.gastos, axis=0)
sd = np.std(df1.gastos, axis=0)
```

```python
#métodos de tratamento de outliers(limite superior e inferior)
#quaisquer valores que estiverem acima ou abaixo desses limites serão considerados anômalos.
mean - 2 * sd 
```

```python
dp = mean + 2 * sd
dp = round (dp , 2)
dp
```

```python
#listando os valores acima da média em duas vezes o desvio padrão
for x in df1.gastos:
    if (x> dp):
        print(x)
```

```python
#separando os valores acima da "média", criando um novo dataframe 
df2 = df1[df1['gastos']> dp]
```

```python
#convertendo o df2 para dataframe
pd.DataFrame(df2)
```

```python
type(df2)
```

```python
#alterando os valores da coluna gastos do df2 para o valor maximo da média em duas vezes o desvio padrão
for i in df2['gastos']:
    df2.gastos = (dp) 
    
df2
```

```python
#removendo os valores atribuidos ao df2
df1 = df1.drop(df1[(df1.gastos >= dp)].index)
```

```python
#argumento concat() para juntar os dois dataframes
df_row = pd.concat([df1, df2])
```

```python
#ordenando por index
df_row.sort_index()
```

```python
#histograma dos gastos 
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
sns.distplot(df_row.gastos);
```

```python
#alterando os valores string sexo e fumantes em boleanos
def transforma_sexo(valor):
    if valor =='mulher':
        return 1
    else:
        return  0
    
    
df_row['sexo_bi'] = df_row['sexo'].map(transforma_sexo)

def fumar(valor):
    if valor == 'sim':
        return 1
    else:
        return  0
    
    
df_row['fumante_bi'] = df_row['fumante'].map(fumar)

df_row.head()
```

```python
#mode o valor mais comum
mode = df_row.gastos.mode()

#a mediana dos valores 
mediana = df_row.gastos.median()

#media
media = df_row.gastos.mean()

#print("valor mais comum é %s\n E da mediana é %r" %mode, %mediana)

print("Valor mais comum é: {}\nA mediana é: {}\nA média é: {:.2f}" .format(mode, mediana, media))
```

# Algoritmo de regressao linear

```python
#algoritmo de regressao linear
lr = LinearRegression()
```

```python
#treinando e gerando modelo 
model = lr.fit(x_train, y_train)
```

```python
#fazendo a predicao(passar o conjunto de dados de teste e armazena para ve o resultado)
resultado = model.predict(x_test)
```

```python
#calculando o erro medio quadratico..é o calculo para ve a media de erro entre o modelo realizado e o valor erro
#para saber se esta performando bem 
mean_squared_error(y_test, resultado) 
```

```python
#porcetagem da variavel resposta
model.score(x_test,y_test)
```

```python
#testando os resultados com conjunto de treino
resultado_treino = model.predict(x_train)
```

```python
mean_squared_error(y_train, resultado_treino)
```

```python
#Ajustando os hiperparamentos do Random Forest
#passamos intervalos de valores e o GridSearch faz testes exastivo e devolve o resultado
from sklearn.model_selection import GridSearchCV
```

```python
parametros = {'min_samples_leaf':[1,10],'min_samples_split':[2,10],'n_estimators':[100,250,500,750]}
```

```python
#atributo para verificar qual a melhor configuracao dos testes para melhores resutlados
grid.best_params_
```

```python
#usando a melhor configuracao
rf_best = grid.best_estimator_
```
